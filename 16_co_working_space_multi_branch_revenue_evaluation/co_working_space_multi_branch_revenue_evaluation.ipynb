#### **The Experimental Sample Data:**

```python
import pandas as pd
import scipy.stats as stats

# Daily revenue samples (in PKR) for the three co-working hubs
blue_area_revenue = [120000, 135000, 142000, 118000, 130000, 125000, 140000]
dha_revenue       = [95000,  105000, 115000, 98000,  110000, 102000, 108000]
gulberg_revenue   = [115000, 122000, 118000, 128000, 120000, 114000, 121000]
```

#### **The Deliverables & Rules:**

##### 1. **Descriptive Baseline Evaluation:** Compute and display the sample mean revenue for each of the three locations separately.

```python
# calculate the sample mean across three locations
blue_area_mean = sum(blue_area_revenue) / len(blue_area_revenue)
dha_mean = sum(dha_revenue) / len(dha_revenue)
gulberg_mean = sum(gulberg_revenue) / len(gulberg_revenue)

print(f"The sample mean revenue for blue area is {blue_area_mean:.2f}")
print(f"The sample mean revenue for dha is {dha_mean:.2f}")
print(f"The sample mean revenue for gulberg is {gulberg_mean:.2f}")
```

##### 2. **Execute the Multi-Group Proof:** Run a One-Way ANOVA test using `scipy.stats`. Extract and print both the $F$-**statistic** and the $p$-**value** clearly.

```python
# run the one-way anova (analysis of variation) test on three independent groups
f_stat, p_value = stats.f_oneway(blue_area_revenue, dha_revenue, gulberg_revenue)

# extract the f-statistics and p-value from the test
print(f"f-statistics is {f_stat:.4f}")
print(f"p-value is {p_value:.4e}")
```

##### 3. **The Executive Decision:** Write a formal `if/else` conditional logic gate using Python that checks your $p$-value against $\alpha = 0.05$. If the null hypothesis is rejected, output an explicit directive explaining what the manager should do next.

```python
# significance value
alpha = 0.05

if p_value < alpha:
    print("\nResult: Statistically Significant!")
    print("""
Blue Area is significantly outperforming the others in terms of avg. daily revenue generation.
So the strategic advice to the Community Operations Manager is to replicate Blue Area's Success and optimizing the underperforming locations!
""")
else:
    print("Result: Not Statistically Significant.")
    print("Small revenue differences are due to random luck!")
```

##### **Post-Hoc Tukey HSD Test:**

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# 1. Combine data into a structured format for Tukey's test
data = (
    blue_area_revenue + dha_revenue + gulberg_revenue
)
labels = (
    ['Blue_Area'] * len(blue_area_revenue) + 
    ['DHA'] * len(dha_revenue) + 
    ['Gulberg'] * len(gulberg_revenue)
)

# 2. Run the pairwise comparison
tukey = pairwise_tukeyhsd(endog=data, groups=labels, alpha=0.05)
print(tukey)
```
