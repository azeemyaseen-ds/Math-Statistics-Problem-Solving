## **The Probability Law:**

Because server crashes are independent events occurring randomly in a fixed interval of time, they follow a **Poisson Distribution**. The probability of experiencing exactly $k$ events is given by the formula:

$$P(X = k) = \frac{\lambda^k \cdot e^{-\lambda}}{k!}$$

Where:

 - $\lambda = 2.4$ (The average rate)
 - $e \approx 2.71828$ (Euler's constant)
 - $k$ is the specific number of crashes you are testing

## **The Deliverables & Rules:**

```python
# import the necessary library for this project
import scipy.stats as stats
```

#### 1. **Perfect Operation Probability:** Calculate the exact probability that the server experiences **zero (0) crashes** next week.

```python
# mu (or lambda) is the historical average rate of events per interval
mu_rate = 2.4

# exact match probability
k_exact = 0
exact_prob = stats.poisson.pmf(k_exact, mu=mu_rate)

#
print(f"Probability of exactly {k_exact} crash is: {exact_prob:.2%}")
```

#### 2. **Standard Threshold Probability:** Calculate the probability that the server experiences **exactly three (3) crashes** next week.

```python
# mu (or lambda) is the historical average rate of events per interval
mu_rate = 2.4

# exact match probability
k_exact = 3
exact_prob = stats.poisson.pmf(k_exact, mu=mu_rate)

print(f"Probability of exactly {k_exact} surges: {exact_prob:.2%}")
```

#### 3. **Catastrophic Failure Risk (Cumulative):** Calculate the probability that the server experiences **more than 4 crashes** next week (i.e., $P(X > 4)$). (Hint: Look into Cumulative Distribution Functions (CDF) or the law of complements).

```python
# mu (or lambda) is the historical average rate of events per interval
mu_rate = 2.4

# cumulative probability (e.g greater than 4)
prob_more_than_4 = stats.poisson.sf(4, mu=mu_rate)

print(f"Probability of getting more than 4 events: {prob_more_than_4:.2%}")
```
