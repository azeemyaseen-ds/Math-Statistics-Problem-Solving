## **The Contingency Matrix Data:**

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Rows: Morning Shift, Evening Shift, Night Shift
# Columns: Low Severity Accidents, High Severity Accidents
# Matrix format: [[Low_Sev, High_Sev], [Low_Sev, High_Sev], ...]
accident_matrix = np.array([
    [45, 10],  # Morning Shift
    [35, 15],  # Evening Shift
    [20, 25]   # Night Shift
])
```

## **The Deliverables & Rules:**

#### 1. **Matrix Framing:** Convert the raw NumPy array into a properly labeled Pandas DataFrame named `df_accidents`, with column headers labeled `['Low_Severity', 'High_Severity']` and index row labels assigned as `['Morning', 'Evening', 'Night']`.

```python
# convert the numpy array into a proper pandas dataframe
df_accidents = pd.DataFrame(accident_matrix)

# rename the columns
df_accidents.rename(columns={0:'Low_Severity', 1:'High_Severity'}, inplace=True)

# assign labels to row index
df_accidents.index = ['Morning', 'Evening', 'Night']
```

```python
# verify the changes
df_accidents
```

#### 2. **Execute the Categorical Proof:** Run the Chi-Square test for independence on your matrix using **scipy.stats.chi2_contingency**. Extract and print the $\chi^2$ **statistic** and the $p$-**value**.

```python
# run the chi-squared test for independence on the warehouse data
chi2, p_value, dof, expected = stats.chi2_contingency(df_accidents)

# Extract & format the result
print(f"The Chi-Squared value is {chi2:.4f}")
print(f"The P-Value is {p_value:.4f}")
```

```python
# # Create a DataFrame of residuals to match your original df_accidents
residual = (accident_matrix - expected) / np.sqrt(expected)

df_residual =  pd.DataFrame(residual, columns=df_accidents.columns, index=df_accidents.index)

print("--- Statistical Residual Matrix ---")
df_residual
```

#### 3. **The Executive Decision:** Code an `if/else` block that compares your $p$-value to $\alpha = 0.05$. Output an explicit operational directive advising the warehouse director whether they need to launch an immediate targeted safety intervention for specific shifts ($H_0$ rejected) or if the distribution is statistically normal and independent ($H_0$ sustained).

```python
# significance value
alpha = 0.05

if p_value < alpha:
    print("Launch an immediate targeted safety intervention for Night Shifts and also increase the supervisory oversight during Night, null hypothesis is rejected!")
else:
    print("Warehouse incidents are completely random and independent across the day, null hypothesis is accepted!")
```
