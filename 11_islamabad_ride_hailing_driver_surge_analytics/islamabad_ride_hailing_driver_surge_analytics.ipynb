## **The Raw Data:**

```python
# import the necessary libraries for the project
import pandas as pd
import numpy as np

# raw data
raw_ride_data = {
    "Hour_Timestamp": ["08:00", "09:00", "10:00", "11:00", "12:00", "13:00", "14:00", "15:00"],
    "F_10_Drivers": [45, 12, 88, 5, 92, 11, 79, 15],
    "I_8_Drivers": [42, 38, 45, 40, 41, 39, 44, 43],
    "D_12_Drivers": [20, 22, 21, 25, 19, 23, 22, 20]
}

# convert the dictionary into a pandas dataframe
df_rides = pd.DataFrame(raw_ride_data)
# view the sample of the data
df_rides.sample(3)
```

```python
# get the overview of the data
df_rides.info()
```

## **The Advanced Business Logic & Rules:**

#### 1. **Central Tendency Computation:** Dynamically calculate the Mean and Median of available drivers for each individual sector (`F_10_Drivers`, `I_8_Drivers`, and `D_12_Drivers`) across the day.

```python
# find the mean and median across each sector
df_rides[['F_10_Drivers', 'I_8_Drivers', 'D_12_Drivers']].agg(['mean', 'median'])
```

#### 2. **Dispersion Diagnosis (The Math Core):** Calculate the **Variance** ($\sigma^2$) and **Standard Deviation** ($\sigma$) for each sector.

```python
# calculate the variance and standard deviation across each sector
df_rides[['F_10_Drivers', 'I_8_Drivers', 'D_12_Drivers']].agg(['var', 'std'])
```

#### 3. **The Coefficient of Variation (CV):** Standard deviations alone can sometimes be misleading if the scales of data are wildly different. Calculate the **Coefficient of Variation (CV)** for each sector as a percentage:$$\text{CV} = \left( \frac{\text{Standard Deviation}}{\text{Mean}} \right) \times 100$$

```python
# calculate the CV % for each sector using the formula;
cv_f_10 = (df_rides['F_10_Drivers'].std() / df_rides['F_10_Drivers'].mean()) * 100
print(f"The Coefficient of Variation of F_10_Drivers is {cv_f_10 / 100:.0%}")

cv_i_8 = (df_rides['I_8_Drivers'].std() / df_rides['I_8_Drivers'].mean()) * 100
print(f"The Coefficient of Variation of I_8_Drivers is {cv_i_8 / 100:.0%}")

cv_d_12 = (df_rides['D_12_Drivers'].std() / df_rides['D_12_Drivers'].mean()) * 100
print(f"The Coefficient of Variation of D_12_Drivers is {cv_d_12 / 100:.0%}")
```
