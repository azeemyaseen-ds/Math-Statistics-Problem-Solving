```python
# import the necessary library for this project
import scipy.stats as stats
```

#### 1. **On-Time Campaign Baseline:** Calculate the exact probability that a randomly chosen delivery takes **less than 35 minutes** (i.e., the percentage of orders that will be successfully delivered on time under the guarantee).

```python
# the avg delivery time of an order
mean = 28.5

# but the order don't deliver exactly at 28.5 mins it is in between 24 mins to 33 mins
std_dev = 4.2

# checking the probability of deliveries taking less than 35 minutes
# cdf => cumulative distribution function
prob_less_than_35 = stats.norm.cdf(35, loc=mean,scale=std_dev)

# final result
print(f"Probality of getting orders under 35 minutes is {prob_less_than_35:.2%}")
```

#### 2. **Financial Risk Exposure:** Calculate the exact probability that a delivery takes **more than 35 minutes** (i.e., the percentage of orders that will fail the guarantee and trigger a free refund).

```python
mean = 28.5
std_dev = 4.2

# checking the probability of deliveries taking more than 35 minutes
prob_more_than_35 = 1 - stats.norm.cdf(35, loc=mean,scale=std_dev)

# final result
print(f"Probality of getting orders more than 35 minutes is {prob_more_than_35:.2%}")
```

#### 3. **The Ultra-Fast Outliers:** Calculate the probability that a delivery is exceptionally fast, taking **between 20 and 25 minutes**. (Hint: Think about how you can subtract a lower cumulative area from an upper cumulative area).

```python
mean = 28.5
std_dev = 4.2

# checking the probability of deliveries between 20 to 25 minutes
prob_between = stats.norm.cdf(25, loc=mean,scale=std_dev) - stats.norm.cdf(20, loc=mean, scale=std_dev)

# final result
print(f"Probability of getting orders between 20 to 25 minutes is {prob_between:.2%}")
```
