## **The Experimental Data:**

```python
import pandas as pd
import numpy as np

dha_real_estate = {
    "Apartment_ID": [101, 102, 103, 104, 105, 106],
    "Actual_Price_Million_PKR": [15.5, 22.0, 18.5, 30.0, 12.0, 26.5],
    "Predicted_Price_Million_PKR": [14.8, 24.2, 17.1, 28.5, 13.5, 25.0]
}

df_properties = pd.DataFrame(dha_real_estate)
```

## **The Deliverables & Rules:**

#### 1. **Residual Computation:** Calculate the raw Residual for each property:
$$\text{Residual} = \text{Actual Price} - \text{Predicted Price}$$
Add this as a new column named `Residual` to your DataFrame.

```python
# compute residual for each property
df_properties['Residual'] = df_properties['Actual_Price_Million_PKR'] - df_properties['Predicted_Price_Million_PKR']

# update the dataset
df_properties
```

#### 2. **Mean Absolute Error (MAE):** Calculate and print the **MAE**, which represents the average magnitude of the errors without considering their direction:
$$\text{MAE} = \frac{1}{n} \sum \vert{} \text{Residual} \vert{}$$

```python
# calculate Mean Absolute Error (average of model mistakes)
mae = np.abs(df_properties['Residual']).mean()

# result
print(f"Mean Absolute Error: {mae:.2f} Million PKR")
```

#### 3. **Mean Squared Error (MSE) & Root Mean Squared Error (RMSE):** Calculate and print the **RMSE**. RMSE penalizes larger errors more heavily, making it a critical metric for financial models:
$$\text{RMSE} = \sqrt{\frac{1}{n} \sum (\text{Residual})^2}$$

```python
# calculate Root Mean Squared Error (find the spread out of errors)
rmse = np.sqrt((df_properties['Residual'] ** 2).mean())

# result
print(f"Root Mean Squared Error: {rmse:.2f} Million PKR")
```
