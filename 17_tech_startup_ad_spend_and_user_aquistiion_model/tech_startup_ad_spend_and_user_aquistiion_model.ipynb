## **The Raw Experimental Data:**

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

growth_data = {
    "Week": [1, 2, 3, 4, 5, 6, 7, 8],
    "Ad_Spend_Thousand_PKR": [12, 18, 25, 30, 40, 45, 55, 60],
    "New_Signups": [150, 210, 310, 340, 490, 530, 640, 690]
}

df_growth = pd.DataFrame(growth_data)
```

## **The Deliverables & Rules:**

#### 1. **Correlation Diagnostics:** Calculate **Pearson's Correlation Coefficient** ($r$) between `Ad_Spend_Thousand_PKR` and `New_Signups`. Print the value and state whether the correlation is strong, moderate, or weak.

```python
# find the pearson's correlation coefficient to check for directional strength
r_value, p_value = stats.pearsonr(df_growth['Ad_Spend_Thousand_PKR'], df_growth['New_Signups'])

print(f"Pearson's r: {r_value:.4f}")
```

**Conclusion:**
 - The correlation is strong and marketing director claim is accurate.
 - The value is close to +1.0, it means simple linear regression model will be highly accurate for forecasting future growth.

#### 2. **Linear Modeling (The Engine):** Use `stats.linregress` to compute the **Slope**, **Intercept**, and $p$-value of the regression line. Print these variables cleanly.

```python
# model the mathematical relationship between ad spend and acquisition
slope, intercept, r_value, p_value, std_err = stats.linregress(
    df_growth['Ad_Spend_Thousand_PKR'],
    df_growth['New_Signups']
)

print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"P-Value: {p_value:.4e}")
```

#### 3. **The Forecasting Engine:** Using the calculated slope and intercept coefficients, compute a new column in your DataFrame named `Predicted_Signups`.

```python
# compute the Predicted_Signups columns using y = m*x + c
df_growth['Predicted_Signups'] = (slope * df_growth['Ad_Spend_Thousand_PKR']) + intercept

# updated dataframe
df_growth
```

#### 4. **The Capital Investment Prediction:** The Director is considering a massive budget boost for Week 9, setting the ad-spend to **85,000 PKR** (which is `85` on our data scale). Calculate and print the exact predicted number of new user sign-ups for that investment level.

```python
# ad spend for week 9
week_9_spend = 85

# predicted user signups based on the ad spend and model variables
predicted_week_9 = (slope * week_9_spend) + intercept

# result
print(f"Predicted New Signups for Week 9: {predicted_week_9:.2f}")
```
